## Functions

```python
import pandas as pd
import numpy as np

import glob
import json

from page_parser import parse_individual_tab
from page_parser import parse_program_tab
from utilities import *

def create_program_tab(program_tab, gen_individual_tab):
    program = {}
    individual_df = pd.DataFrame(gen_individual_tab)
    splitted_name = program_tab.pop('program').split(',')
    for i in range(len(splitted_name)):
        splitted_name[i] = splitted_name[i].strip()
    program['uni'], program['dept'], program['prog'] = splitted_name[0], splitted_name[1], splitted_name[2]
    if program_tab['historical']:
        program['year'] = 2018
    else:
        program['year'] = int(program_tab['file'].split('/')[1][:4])
    program['observation_date'] = pd.to_datetime(
        program_tab['file'].split('/')[1], format='%Y%m%d'
    ).strftime('%Y-%m-%d')
    if 'атт' in individual_df:
        orig = dummy(individual_df, 'атт', 'ориг')
        bvi = dummy(individual_df, 'тип', 'бви')
        orig_not_bvi = [bool(orig[i]*(1-bvi[i])) for i in range(individual_df.shape[0])]
        program['n_exams'] = len(find_exams_colnames(individual_df))
        program['mean_sum'] = sum(individual_df.loc[orig_not_bvi, find_sum_colname_max(individual_df)[0]].values)/sum(orig_not_bvi)
        program.update(create_exams_dict(individual_df, orig_not_bvi))
        program['атт_check'] = sum(orig)
        program['бви_check'] = sum(bvi)
        
    program.update(program_tab)
    
    return program

def parse_page(page):
    return create_program_tab(
        parse_program_tab(page),
        parse_individual_tab(page)
    )
    
    
with open('output.json', 'w') as f: # add date!
    f.write('[')
    try:
        for page in glob.iglob('pages' + '/**/*.html', recursive=True):
            if not 'index.html' in page:
                page = parse_page(page)
                f.write(json.dumps(page))
                f.write(',\n')
    finally:
        f.write(']')
```
